```python
import joblib
import numpy as np


amass = joblib.load('data/vibe_db/amass_db.pt')

for k in amass.keys():
    print(k, amass[k].shape)

def merge_datasets(dataset_1, dataset_2):
    # Check if both datasets have the same keys and compatible shapes
    if dataset_1.keys() != dataset_2.keys():
        raise ValueError(
            "Datasets have different keys and cannot be merged")

    # Merge the second dataset into the first dataset
    for key in dataset_1.keys():
        # Make sure the shapes are compatible, excluding the first dimension
        if dataset_1[key].shape[1:] != dataset_2[key].shape[1:]:
            raise ValueError(
                f"Datasets are not compatible for the key: {key}")

        # Append the data from the second dataset to the first dataset
        dataset_1[key] = np.concatenate(
            (dataset_1[key], dataset_2[key]), axis=0)

    # Save the merged dataset back to a .pt file
    joblib.dump(dataset_1, 'data/vibe_db/amass_merged.pt')

    del dataset_1


merge_datasets(amass, joblib.load('data/vibe_db/amass_grab_weizmann_db.pt'))

amass_merged = joblib.load('data/vibe_db/amass_merged.pt')

for k in amass_merged.keys():
    print(k, amass_merged[k].shape)
```
